```python
%matplotlib inline
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
#正規分布に従った確率秘度関数を生成してグラフを描画する
# xは。ー３から＋３まで、0.01ステップで生成したデータ
x = np.arange(-3,3,0.001) #平均＝0、標準偏差＝１の正規分布に従う確率関数

y = norm.pdf(x)

plt.plot(x,y) #データをプロットする
plt.show()  #グラフを画面に表示する
```

```python
## 複数のグラフを描画
```

```python
plt.plot(x,y)#平均＝0、標準偏差＝0.5の正規分布に従う確率関数
y1 = norm.pdf(x,1.0,0.5)
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y1)
plt.savefig('./myplot.png',format='png')# グラフをファイルに保存し、pathを指定する。
```
